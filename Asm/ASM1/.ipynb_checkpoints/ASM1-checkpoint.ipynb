```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder

from matplotlib import style
style.use('dark_background')
```

```python
trainData = pd.read_csv("train_data.csv"
                        , delimiter= ",")
trainData = trainData.drop(['ID'], axis=1)
trainData = trainData.drop(['HealthServiceArea'], axis=1)
testData = pd.read_csv("test_data.csv"
                      , delimiter = ",")
trainData.head(20)
```

```python
# Transform LengthOfStay into binary class (0 for <= 4 and 1 for > 3 days)
trainData['LengthOfStay'] = trainData['LengthOfStay'].apply(lambda x: 0 if int(x) <= 3 else 1)
```

```python
def snsFacetGrid(df):
    for c in df.columns:
        g = sns.FacetGrid(df, col='LengthOfStay',height =4, aspect = 2)
        g.map(plt.hist, str(c) ,bins=20)
        g.set_axis_labels(c, "count")
```

```python
# apply string index for TypeOfAdmission column for classification
typeOfAdmissionIndex = {'Newborn' : 1, 'Urgent' : 2, 'Emergency' : 3, 'Elective' : 4}

trainData['TypeOfAdmission'] = trainData['TypeOfAdmission'].apply(lambda x : typeOfAdmissionIndex[x])
```

```python
# apply binary classification for EmergencyDepartmentIndicator
trainData['EmergencyDepartmentIndicator'] = trainData['EmergencyDepartmentIndicator'].apply(lambda x : 1 if x == 'Y' else 0)
```

```python
# perform dummies variables approach for column Race

# creates new dataframe with columns for each race
# and map the Race values to its corresponding column
dummiesDF = pd.get_dummies(trainData.Race)

# merge the new dataframe to the orginal
trainData = pd.concat([trainData, dummiesDF], axis='columns')

# drop column PaymentTypology and the last column ('Other Race') (for better practice)
trainData = trainData.drop(['Race', 'Other Race'], axis='columns')
```

```python
# perform dummies variables approach for column PaymentTypology

# create new dataframe with columns for each type of payment
# and map the PaymentTypology values to its corresponding column
dummiesDF = pd.get_dummies(trainData.PaymentTypology)

# merge the new dataframe into the orginal
trainData = pd.concat([trainData, dummiesDF], axis='columns')

# drop column PaymentTypology and the last column ('Unknown') (for better practice)
trainData = trainData.drop(['PaymentTypology', 'Unknown'], axis='columns')
```

```python
# perform dummies variables approach for column Gender

# create new dataframe for each type of gender
# and map the PaymentTypology values to its corresponding column
dummiesDF = pd.get_dummies(trainData.Gender)

# merge the new dataframe into the orginal
trainData = pd.concat([trainData, dummiesDF], axis='columns')

# drop column Gender and the last column ('U') (for better practive)
trainData = trainData.drop(['Gender', 'U'], axis='columns')
```

```python
trainData.head(20)
```

### Multivariable Regression

```python
# split data into train and test set
from sklearn.model_selection import train_test_split

X = trainData.drop(['LengthOfStay'], axis='columns')
Y = trainData[['LengthOfStay']]

trainX, testX, trainY, testY = train_test_split(np.array(X),
                                               np.array(Y),
                                               test_size = 0.25)

print(trainX.shape)
print(testX.shape)
print(trainY.shape)
print(testY.shape)
```

```python
# perform logistic regression
from sklearn import linear_model

# create model instance
logReg = linear_model.LogisticRegression(max_iter = 4000)
```

```python
# fit data into model
logReg.fit(trainX, trainY.ravel())
```

```python
from sklearn.metrics import mean_squared_error
# make prediction with created test set
prediction = logReg.predict(testX)
print('Mean squared error ', mean_squared_error(testY, prediction))
```

```python
score = logReg.score(testX, testY)
print("Accuracy: ", score*100, end="")
print("%")
```
