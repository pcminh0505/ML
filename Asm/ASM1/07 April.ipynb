```python
for col in X.columns:
    print(col)
    count = X[col].value_counts()
    percent = X[col].value_counts(normalize=True).mul(100).round(2).astype(str) + '%'
    print(pd.DataFrame({'Counts': count, 'Percent': percent}))
    print('--------------------------------')
```
